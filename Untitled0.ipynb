```python
import pandas as pd
import numpy as np

def generate_synthetic_data(filename, is_faulty=False, duration_sec=10, sample_rate=100):
    t = np.linspace(0, duration_sec, duration_sec * sample_rate)

    # Base Motor Frequency (50Hz vibration)
    base_vibration = 0.5 * np.sin(2 * np.pi * 50 * t)

    if is_faulty:
        # Add a low-frequency "imbalance" wobble (10Hz)
        fault_vibration = 1.2 * np.sin(2 * np.pi * 10 * t)
        # Higher random noise for faulty state
        noise = np.random.normal(0, 0.5, len(t))
        z_axis = base_vibration + fault_vibration + noise + 9.81 # + Gravity
    else:
        # Lower noise for healthy state
        noise = np.random.normal(0, 0.1, len(t))
        z_axis = base_vibration + noise + 9.81 # + Gravity

    # Simulate X and Y axes with minor jitter
    x_axis = np.random.normal(0, 0.05, len(t))
    y_axis = np.random.normal(0, 0.05, len(t))

    df = pd.DataFrame({'accX': x_axis, 'accY': y_axis, 'accZ': z_axis})
    df.to_csv(filename, index=False, header=False)
    print(f"Generated: {filename}")

# Create the files
generate_synthetic_data('healthy.csv', is_faulty=False)
generate_synthetic_data('faulty.csv', is_faulty=True)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

# 1. Load your raw data (Upload your 'healthy.csv' and 'faulty.csv' to Colab)
def process_vibration(file_path, label):
    df = pd.read_csv(file_path, names=['accX', 'accY', 'accZ'])

    # 2. Perform FFT (Converting Time to Frequency)
    # Assuming 100Hz sampling rate
    N = len(df)
    T = 1.0 / 100.0
    yf = fft(df['accZ'].values) # Analyzing Vertical vibration
    xf = fftfreq(N, T)[:N//2]

    plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
    plt.title(f"Vibration Spectrum: {label}")
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Amplitude")
    plt.show()

# Run the analysis
process_vibration('healthy.csv', 'Healthy Motor')
process_vibration('faulty.csv', 'Faulty Motor')
```

```python
import tensorflow as tf
from tensorflow.keras import layers

def build_custom_model(input_shape):
    model = tf.keras.Sequential([
        # Layer 1: Convolution (Extracting vibration patterns)
        layers.Conv1D(filters=16, kernel_size=3, activation='relu', input_shape=input_shape),
        layers.MaxPooling1D(pool_size=2),

        # Layer 2: Flattening for Classification
        layers.Flatten(),
        layers.Dense(16, activation='relu'),

        # Output Layer: Softmax for probability [Healthy, Faulty]
        layers.Dense(2, activation='softmax')
    ])

    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

print("Custom Research Model Architecture Defined.")
```

```python
import matplotlib.pyplot as plt
from scipy.fft import fft, fftfreq

def plot_fft(filename, title):
    df = pd.read_csv(filename, names=['accX', 'accY', 'accZ'])
    signal = df['accZ'].values - 9.81 # Remove gravity

    N = len(signal)
    yf = fft(signal)
    xf = fftfreq(N, 1/100)[:N//2] # 100Hz sample rate

    plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
    plt.title(title)
    plt.xlabel("Frequency (Hz)")
    plt.ylabel("Magnitude")
    plt.grid()

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1); plot_fft('healthy.csv', 'Healthy: Clean 50Hz Peak')
plt.subplot(1, 2, 2); plot_fft('faulty.csv', 'Faulty: 10Hz Imbalance Detected')
plt.tight_layout()
plt.show()
```

```python
def create_windows(filename, window_size=100):
    df = pd.read_csv(filename, names=['accX', 'accY', 'accZ'])
    data = df[['accX', 'accY', 'accZ']].values

    windows = []
    # Slide a window across the data
    for i in range(0, len(data) - window_size, window_size // 2):
        windows.append(data[i : i + window_size])

    return np.array(windows)

# Create training sets
X_healthy = create_windows('healthy.csv')
X_faulty = create_windows('faulty.csv')

# Combine and create labels (0 = Healthy, 1 = Faulty)
X = np.concatenate([X_healthy, X_faulty])
y = np.concatenate([np.zeros(len(X_healthy)), np.ones(len(X_faulty))])

# Shuffle the research data
idx = np.random.permutation(len(X))
X, y = X[idx], y[idx]
```

```python
from tensorflow.keras import layers, models

model = models.Sequential([
    # Layer 1: Conv1D acts as a "Digital Filter" to find vibration patterns
    layers.Conv1D(filters=16, kernel_size=3, activation='relu', input_shape=(100, 3)),
    layers.MaxPooling1D(pool_size=2),

    # Layer 2: Deeper feature extraction
    layers.Conv1D(filters=32, kernel_size=3, activation='relu'),
    layers.GlobalAveragePooling1D(),

    # Output Layer: [Probability of Healthy, Probability of Faulty]
    layers.Dense(2, activation='softmax')
])

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Train the model
history = model.fit(X, y, epochs=20, validation_split=0.2)
```
